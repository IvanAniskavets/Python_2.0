## Задание 2
Непрерывная случайная величина задана плотностью распределения:$$ f(\xi) = C, \xi \in [0;5];f(\xi) = 0, \xi \notin [0;5]$$Найдите C, математическое ожидание $\xi$ а также вероятность попадания в отрезок $[3.5;7]$, т.е. P(3.5 < $\xi$ < 7).



```python
a = 0
b = 5

uni = sts.uniform(a, b-a)
x = np.linspace(-5, 10, 1000)
pdf = uni.pdf(x)
plt.plot(x, pdf)
plt.title('Плотность вероятности функции из заданных')
plt.show()

if 0.2 == uni.pdf(np.random.uniform(a, b)):
    print('Плотность вероятности имеет максимум в точке 0.2')
else:
    print('Максимум плотности вероятности найден неверно')
```

```python
if 2.5 == scipy.integrate.quad(lambda x: x*1/5, 0, 5)[0]:
    print('Математическое ожидание расчитано верно')
else:
    print('Математическое ожидание расчитано неверно')
```

```python
if 0.3 == round(scipy.integrate.quad(lambda x: 1/5, 3.5, 5)[0], 1):
    print('Вероятность попадания в интервал расчитана верно')
else:
    print('Вероятность попадания в интервал расчитана неверно')
```

## Задание 3
Известно, что светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным. Посчитайте, какое количество информации несёт сообщение о цвете светофора в текущий момент. А что будет в случае 28, 28 и 28 секунд для каждого цвета?

```python
# первый неравновероятный случай

t = 54 + 4 + 28 # общее время светофора

p1 = 54 /t # вероятность красного цвета
p2 = 4 /t  # вероятность желтого цвета
p3 = 28/t  # вероятность зеленого цвета

I1 = - math.log2(p1)
I2 = - math.log2(p2)
I3 = - math.log2(p3)

Shannon = np.abs((-(p1*I1)) + (-(p2*I2)) + (-(p3*I3)))

Shannon
```
