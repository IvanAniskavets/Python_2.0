```python
import numpy as np
from scipy.misc import derivative
import matplotlib.pyplot as plt
from sympy import diff, symbols, cos, sin
```

### Task 1

```python
x, y = symbols('x y')
function = x**5 + 4 * sin(2 * x) + cos( 3 * x + 3)
dif_1 = diff(function, x)
dif_2 = diff(diff(function, x))
print("Первая производная : {}".format(dif_1))
print("Вторая производная : {}".format(dif_2))
```

```python
x = symbols('x')
dif_1 = diff(function, x).evalf(subs={'x':1})
dif_2 = diff(diff(function, x)).evalf(subs={'x':1})
print("Значение первой производная : {}".format(dif_1))
print("Значение второй производная : {}".format(dif_2))
```

```python
x, y = symbols('x y')
x = 1
function = x**5 + 4 * np.sin(2 * x) + np.cos( 3 * x + 3)

print(function)
```

```python
# функция

def f(x):
    return x**5 + 4 * np.sin(2 * x) + np.cos( 3 * x + 3)
```

```python
x = 1.

print("Значение первой производная : {}".format(derivative(f, x, dx=1e-6))) # первая производная
print("Значение второй производная : {}".format(derivative(f, x, dx=1e-6, n=2))) # вторая производная
```

```python
def f(x):
    return np.sin(2*x + 1)**5

x_array = np.linspace(-5., 5., 500)

print("Shape x:", x_array.shape)
print('========================================================')

fx = f(x)

y = np.array([f(x) for x in x_array])
yp = np.array([derivative(f, xx, dx=1e-6) for xx in x_array])
ypp = np.array([derivative(f, xx, dx=1e-6, n=2) for xx in x_array])


print("Shape y:", y.shape)
print("Shape yp:", yp.shape)
print("Shape ypp:", ypp.shape)
```

```python
plt.plot(x_array, y, color='orange', label='function')
plt.plot(x_array, yp, color='b', label="derivative_1'")
plt.plot(x_array, ypp, color='r', label="derivative_2''")
plt.grid(True)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
def f(x):
    return np.sin(2*x + 1)**5

x = np.linspace(-5, 5)
fx = f(x)
f1x = [derivative(f, xx, dx=1e-6 , n=1 ) for xx in x]
f2x = [derivative(f, xx, dx=1e-6 , n=2 ) for xx in x]
plt.plot(x, fx, color='orange', label = 'function' )
plt.plot(x, f1x, color='b', label = 'derivative_1' )
plt.plot(x, f2x, color='r', label = 'derivative_2' )
plt.grid(True)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```
