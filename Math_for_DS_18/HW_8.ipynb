```python
import math 
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import scipy.integrate
```

```python
#  1 
#   2, 3, 5, 7, 13

total_P = 0 #    
n = 15 #   
p = 0.5 #   
q = 1 - p #    
k = 2 # -  

#  
# C = math.factorial(n) / (math.factorial(k) * math.factorial(n - k))
# P = C * p**k *(1 - p)**(n-k)

for k in [2, 3, 5, 7, 11, 13]:   # k - -  
    C = math.factorial(n) / (math.factorial(k) * math.factorial(n - k))
    P = C * p**k *(1 - p)**(n-k)
    total_P += P


print(f'      15   {round(total_P * 100, 0)} %')
```

```python
#  2


for vals in [(2.0, 0.5), (1.5, 0.5), (2.5, 0.5), (2.0, 0.45), (2.0, 0.55)]:
    norm_rv = sts.norm(loc=vals[0], scale=vals[1])
    x = np.linspace(0, 4, 100)
    y = norm_rv.pdf(x)
    plt.plot(x, y, label=f" = {vals[0]},  = {vals[1]}")

plt.title(" ")
plt.ylabel("f(x)")
plt.xlabel("x")
plt.legend()
plt.grid()
plt.show()
```

```python
for lamda in [0.1, 0.5, 0.9, 1.1, 1.5]:
    expon_rv = sts.expon(lamda)
    x = np.linspace(0, 5, 1000)
    y = expon_rv.pdf(x)
    plt.plot(x, y, label=f" = {lamda}")

plt.title(" ")
plt.ylabel("f(x)")
plt.xlabel("x")
plt.legend()
plt.grid()
plt.show()
```

  ,   .

```python
for n in [1, 2, 3, 5, 30]:
    t_rv = sts.t(n)
    x = np.linspace(-5, 5, 1000)
    y = t_rv.pdf(x)
    plt.plot(x, y, label=f"n = {n}")

plt.title(" ")
plt.ylabel("f(x)")
plt.xlabel("x")
plt.legend()
plt.grid()
plt.show()
```
