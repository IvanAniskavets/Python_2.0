# Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
# читаю данные с оценками юзеров и фильтруем данные оставляя только фильмы с оценкой 5.0

df_rating = pd.read_csv('C:/Netology/ml-25m/ratings.csv')
df_rating = df_rating.query('rating == 5.0')
df_rating.head()
```

```python
# читаю данные с названиями и жанрами фильмов

df_movies = pd.read_csv('C:/Netology/ml-25m/movies.csv')
df_movies.head()
```

```python
# добавляю название фильма к данным с оценками юзеров

df_rating = df_rating.merge(df_movies[['title', 'movieId']], on='movieId', how='left')
df_rating.head()
```

```python
# применяю фильтр к оценкам, оставляем оценки равные 5.0 , применяем метод value_counts() и выводим 1 значение

df_rating[['title', 'rating']].value_counts().head(1)
```

### Ответ:
Фильму Shawshank Redemption было выставлено больше всего оценок 5.0

# Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
# читаем данные

countries = pd.read_csv('C:/Netology/Lecture_notebook/10._Basics_Pandas/power.csv')
countries.head()
```

```python
# фильтруем и выбираем данные только по Латвии, Литве и Эстонии

pribaltica_countries = countries.query('country == "Latvia" or country == "Lithuania" or country == "Estonia"')
pribaltica_countries.head()
```

```python
# фильтр датафрейма стран Прибалтики с учетом категорий, вывод только категорий 4, 12 и 21
# фильтр датафрейма стран Прибалтики с учетом положительного значение quantity (quantity = 0 или quantity > 0)
# фильтр датафрейма стран Прибалтики с года (с 2005 по 2010 включительно)

filtered_baltics_countries = pribaltica_countries.query('category == 4 or category == 12 or category == 21').query('2005 <= year <= 2010 and quantity >= 0')
filtered_baltics_countries.head(5)
```

```python
# суммирование столбца quantity
# группировка по столбцам country, year, category
# сортировка по year, country, category

filtered_baltics_countries.groupby(by=['country','year', 'category']).sum().sort_values(by=['year', 'country', 'category'], ascending=[True, True, True]).head(60)
```
