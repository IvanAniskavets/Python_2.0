# Визуализация № 1

```python
# визуализируем распределения наблюдений в наборе данных, аналогичное гистограмме

fig = plt.figure(figsize=(7, 5))
ax = sns.kdeplot(x='Ladder_score', color='#FF9001', fill=True, data=df)
ax.set_title('График распределения оценок', fontdict={'fontsize':18}, pad=16)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.grid()
ax.set_xlabel('Оценка')
ax.set_ylabel('Плотность');
```

```python
ax = sns.distplot(df.Ladder_score)
ax.set_title('Распределение оценок')
ax.grid()
ax.set_xlabel('Оценка')
ax.set_ylabel('Плотность');
```

# Визуализация № 3
<a id='task4'></a>

```python
# рисуем гистограмму из n стран с верхними и нижними значениями

n = 10

temp = df.sort_values('Ladder_score', ascending=False)
data = pd.concat([temp[:n], temp[-n:]])[['Country_name', 'Regional_indicator', 'Ladder_score']]
data_regions = df['Regional_indicator'].reset_index(drop=True)
    
palette = list(data_regions.map(lambda x: color_dict[x]))
    

fig = plt.figure(figsize=(13,7))
    
ax = sns.barplot(x='Ladder_score', y='Country_name', palette=palette, orient='h', data=data)
    
for i in range(n):
    ax.annotate(data_regions[i], xy=(0.25, i+0.15), size=10, weight='bold', color='#FFFFFF')
    
for i in range(n, n*2):
    ax.annotate(data_regions[i], xy=(0.25, i+0.15), size=10, weight='bold', color='#FFFFFF')

ax.set_title(f'Оценка: Самые счастливые VS. Аутсайдеры {n}', fontdict={'fontsize':18}, pad=16)
ax.set_xlabel('Оценка')
ax.set_ylabel('');
    
```

### Вывод:
<a id='subtask3'></a>
Теперь, более подробно мы можем увидеть рейтинг самых счастливых стран в разрезе регионов в сравнении со странами с худшими показателями. 
Как видим, самыми счастливыми странами с уверенностью можно назвать страны Скандинавии.

# Визуализация № 4
<a id='task5'></a>

```python
# группируем считаем средние показатели, 
# для этого используем агрегирующую функцию mean() библиотеки Pandas и группируем данные по регионам

group = df.groupby('Regional_indicator').mean()
group = pd.DataFrame(group).reset_index()
```

```python
def pair_plot(x, title_x, xlim, x_label_pos, y_label_pos):
    y, ylim = 'Ladder_score', (2.2, 8.2)
    fig, axes = plt.subplots(1,2, figsize=(17,6))

    ax0 = axes[0]
    sns.scatterplot(x=x, y=y, data=df, color=main_color, alpha=0.7, ax=ax0)
    sns.regplot(x=x, y=y, data=df, color=main_color, ax=ax0, scatter=False)
    ax0.set_xlim(xlim)
    ax0.set_ylim(ylim)
    ax0.spines['right'].set_visible(False)
    ax0.spines['top'].set_visible(False)
    ax0.set_xlabel(title_x)
    ax0.set_ylabel('Оценка')
    
    ax1 = axes[1]
    sns.scatterplot(x=x, y=y, data=group, color=main_color, ax=ax1)
    for region in group.Regional_indicator:
        plt.text(x=group[x][group.Regional_indicator==region]+x_label_pos, 
                 y=group[y][group.Regional_indicator==region]+y_label_pos, 
                 s=region,
                 fontdict=dict(alpha=0.8, size=7.5))
    
    sns.regplot(x=x, y=y, data=df, color=main_color, ax=ax1, scatter=False)
    ax1.set_xlim(xlim)
    ax1.set_ylim(ylim)
    ax1.spines['right'].set_visible(False)
    ax1.spines['top'].set_visible(False)
    ax1.set_xlabel(title_x)
    ax1.set_ylabel('Оценка')
    
    fig.suptitle(f'Зависимость показателя {title_x} к Оценке', fontsize=15, y=0.95);
```

```python
x = 'Logged_GDP_per_capita'
title_x = 'Зарегистрированного ВВП на душу населения'

pair_plot(x, title_x, (6.2,12.2), 0.07, -0.03)
```

```python
x = 'Social_support'
title_x = 'Социальной поддержка'

pair_plot(x, title_x, (0.4, 1), 0.01, -0.05)
```

```python
x = 'Healthy_life_expectancy'
title_x = 'Продолжительности жизни'

pair_plot(x, title_x, (45,80), 0.5, -0.02)
```

```python
x = 'Freedom_to_make_life_choices'
title_x = 'Свободы делать жизненный выбор'
pair_plot(x, title_x, (0.3,1), 0.007, -0.02)
```

```python
x = 'Perceptions_of_corruption'
title_x = 'Восприятия коррупции'
pair_plot(x, title_x, (0,1), 0.01, -0.02)
```

К таким факторам относится ВВП на душу населения, Социальная поддержкаб Продолжительность жизни, Свобода делать жизненный выбор, Восприятие коррупции.
