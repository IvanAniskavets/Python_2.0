```python
# задание 1 
# простые числа 2, 3, 5, 7, 13

total_P = 0 # сумма вероятностей интересующих исходов
n = 15 # число независимых испытаний
p = 0.5 # вероятность выпадения орла
q = 1 - p # вероятно не выпадения орла
k = 2 # кол-во благоприятных исходов

# формула Бернулли
# C = math.factorial(n) / (math.factorial(k) * math.factorial(n - k))
# P = C * p**k *(1 - p)**(n-k)

for k in [2, 3, 5, 7, 13]:   # k - кол-во благоприятных исходов
    C = math.factorial(n) / (math.factorial(k) * math.factorial(n - k))
    P = C * p**k *(1 - p)**(n-k)
    total_P += P

print(f'Вероятность выпадения простого числа орлов равна {round(total_P, 4)}.')
```

Задание 4

Обозначим через В событие, заключающееся в том, что будет куплена продукция высшего сорта, через  А1, А2, А3 обозначим события, заключающиеся в покупке продукции, принадлежащей соответственно первому, второму и третьему предприятиям.
Можно применить формулу полной вероятности, причем в наших обозначениях:  
P(A1) = 0.2        P(B|A1) = 0.1  
P(A2) = 0.3        P(B|A2) = 0.05  
P(A3) = 0.5       P(B|A3) = 0.2  
Подставляем эти значения в формулу полной вероятности:  
P(B) = 0.2 * 0.1 + 0.3 * 0.05 + 0.5 * 0.2 = 0.135

Вероятность того, что случайно купленная новая продукция окажется высшего сорта равна 0.135
