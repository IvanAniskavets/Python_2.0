```python
# find  / findall функции bs4 ищит инфу на странице, article это тег, post - это класс, 
# всегда возвращает список, в отличии от find

all_info = soup_2.find_all('article', class_='post')
all_info
```

```python
# добрались до ссылок

list_link = []
for article in all_info:
    my_link = article.find('a', class_= 'post__title_link').get('href')
    list_link.append(my_link)
    
list_link
```

```python
# обернем наши действия в функцию , только одна страница

def all_links_1(URL):
    res = requests.get(URL)
    soup = BeautifulSoup(res.text, 'html.parser')
    all_info = soup.find_all('article', class_='post')
    list_link = []
    for article in all_info:
        my_link = article.find('a', class_= 'post__title_link').get('href')
        list_link.append(my_link)
    return list_link
```

```python
# обернем наши действия в функцию , все страницы с постами

def all_links_2(URL):
    list_link = []
    res_1 = requests.get(URL)
    soup_1 = BeautifulSoup(res_1.text, 'html.parser')
    number_page = soup_1.find('a', class_='toggle-menu__item-link toggle-menu__item-link_pagination toggle-menu__item-link_bordered').get('href')
    for x in range(1,int(number_page[-3:-1])+1):
        url_links = f'https://habr.com/ru/all/page{x}'
        res_2 = requests.get(url_links)
        soup_2 = BeautifulSoup(res_2.text, 'html.parser')
        all_info = soup_2.find_all('article', class_='post')
        for article in all_info:
            my_link = article.find('a', class_= 'post__title_link').get('href')
            list_link.append(my_link)
    return list_link
```

```python
# присваиваем переменной значение списка со всеми ссылками на посты
res_all_links = all_links_2(URL)
res_all_links
```

## Решение: итоговый датафрейм
<a id='subtask1'></a>

```python
# формируем итоговый дата-фрейм

our_table = pd.DataFrame()

for el in res_all_links:
    time.sleep(0.36)
    res_res = requests.get(el)
    res_soup = BeautifulSoup(res_res.text, 'html.parser')
    text_post = res_soup.find('article', class_='post') # в этой строке лежит HTML 
    for word in words:
        if word.lower() in text_post.text.lower():
            title_post = text_post.find('span', class_='post__title-text').text
            date_post = text_post.find('span', class_='post__time').text
            post_text = text_post.find('div', class_='post__text').text
            #link_post = text_post.find('div', class_='post__text').get('data-io-article-url')
            row = {'date': date_post, 'title': title_post, 'link': el, 'text': post_text}
            our_table = pd.concat([our_table, pd.DataFrame([row])]) # в пустой датафрейм добавляем каждую новую строку с данными
            break
        
our_table 
```
