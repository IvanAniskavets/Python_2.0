```python
import numpy as np
import pandas as pd
import seaborn as sns
from random import randint

from sklearn.linear_model import LinearRegression
```

```python
import random
```

```python
# генерирую данные для разных типов изделия

class_1 = np.random.choice([0], 787) 
class_2 = np.random.choice([1], 163) 
class_3 = np.random.choice([2], 50) 

classes = np.concatenate((class_1 , class_2, class_3))

random.shuffle(classes)

classes
```

```python
# Создаём сэмпл
n_samples = 1000

product_type = classes
product_weight_kg = np.linspace(0.1, 0.9, n_samples) * 100000
product_length_m = np.random.choice(5, n_samples) + 2
lifetime_years = np.random.choice(10, n_samples)
product_reliability_level = np.random.choice(100, n_samples) / 100

multiplying_factor = lifetime_years * product_reliability_level * 0.1 + 0.3


data = pd.DataFrame({'product_type': product_type, 'product_weight_kg': product_weight_kg, 'product_length_m': product_length_m, 
                     'lifetime_years': lifetime_years, 'product_reliability_level': product_reliability_level, 'multiplying_factor': multiplying_factor})
```

```python
data.head(5)
```

```python
# Создаем новый признак

data.loc[(data.product_type == 0), 'tax'] = 0.18
data.loc[(data.product_type == 1), 'tax'] = 0.15
data.loc[(data.product_type == 2), 'tax'] = 0.30
```

```python
data.sample(15)
```

```python
# Создаем ключевую метрику, прайс на изделие, 1 кг изделия = 15 $ 

data['price, $'] = ((data['product_weight_kg'] * 15 + data['product_length_m']) * (data['lifetime_years'] * (data['product_reliability_level']+data['multiplying_factor']))) * (1 + data['tax'])
```

```python
data.head(15)
```

```python
data.columns
```

```python
from sklearn.metrics import mean_absolute_error


X = data[['product_weight_kg', 'product_length_m',
       'lifetime_years', 'product_reliability_level', 'multiplying_factor',
       'tax']]
y = data['price, $']

reg = LinearRegression().fit(X, y)

print('Weights: {}'.format(reg.coef_))
print('Bias: {}'.format(reg.intercept_))

pred_values = reg.predict(X)

print('Error: {}'.format(mean_absolute_error(pred_values, y)))
```
