Затем мы создаем корреляционную матрицу, которая измеряет линейные отношения между переменными. Корреляционная матрица может быть сформирована с помощью функции corr из библиотеки pandas dataframe. Мы будем использовать функцию тепловой карты из библиотеки seaborn для построения корреляционной матрицы.

```python
correlation_matrix = data.corr().round(2)

# annot = True для печати значений внутри квадрата

sns.heatmap(data=correlation_matrix, annot=True);
```

Коэффициент корреляции варьируется от -1 до 1. Если значение близко к 1, это означает, что существует сильная положительная корреляция между двумя переменными. Когда он близок к -1, переменные имеют сильную отрицательную корреляцию.

Выводы:
* Чтобы соответствовать модели линейной регрессии, мы выбираем те характеристики, которые имеют высокую корреляцию с нашей целевой переменной MEDV. Посмотрев на матрицу корреляции, мы можем увидеть, что RM имеет сильную положительную корреляцию с MEDV (0,7), тогда как LSTAT имеет высокую отрицательную корреляцию с MEDV (-0,74).  
* Важным моментом при выборе функций для модели линейной регрессии является проверка мультиколинейности. Характеристики RAD, TAX имеют соотношение 0,91. Эти пары функций сильно коррелированы друг с другом. Мы не должны выбирать обе эти функции вместе для обучения модели. Проверьте это для объяснения. То же самое касается функций DIS и AGE, у которых соотношение составляет -0,75.

Основываясь на приведенных выше наблюдениях, мы будем использовать RM и LSTAT в качестве наших функций. Используя диаграмму рассеяния, давайте посмотрим, как эти функции меняются в зависимости от MEDV.

```python
plt.figure(figsize=(20, 5))

features = ['LSTAT', 'RM']
target = data['MEDV']

for i, col in enumerate(features):
    plt.subplot(1, len(features) , i+1)
    x = data[col]
    y = target
    plt.scatter(x, y, marker='o')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('MEDV')
```

Построим прогноз y^ на основе LSTAT    
Здесь мы имеем β0 (Взято из таблицы доверительного интервала) = 31,6173 и β1 = -0,5520 (коэффициент LSTAT). Теперь сделаем модель из этих входных данных. то есть:   
y_hat = β0 + β1 * LSTAT = 31,6173 + (- 0,5520) * LSTAT

```python
LSTAT = data["LSTAT"] 
plt.scatter(LSTAT,data["MEDV"])
y_hat=31.6173+(-0.5520)*LSTAT
plt.plot(LSTAT,y_hat,lw=5,c="green")
plt.xlabel("LSTAT",fontsize=20)
plt.ylabel("MEDV",fontsize=20);
```

Выводы:
* Цены растут по мере линейного увеличения стоимости RM. Есть несколько выбросов, а количество данных, похоже, ограничено 50.  
* Цены имеют тенденцию к снижению с увеличением LSTAT. Хотя это не похоже на то, чтобы следовать точно по линейной линии.

Подготовка данных для обучения модели
Мы объединяем столбцы LSTAT и RM, используя np.c_, предоставленный библиотекой numpy.

```python
X = pd.DataFrame(np.c_[data['LSTAT'], data['RM']], columns = ['LSTAT','RM'])
Y = data['MEDV']
```

Разделение данных на обучающие и тестовые наборы  
Затем мы разбиваем данные на обучающие и тестовые наборы. Мы обучаем модель на 80% образцов и тестируем на оставшихся 20%. Мы делаем это, чтобы оценить эффективность модели на невидимых данных. Для разделения данных мы используем функцию train_test_split, предоставленную библиотекой scikit-learn. Наконец, мы печатаем размеры нашего обучающего и тестового набора, чтобы проверить, правильно ли произошло разделение.

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

Обучение и тестирование модели  
Мы используем LinearRegression scikit-learn для обучения нашей модели как на обучающих, так и на тестовых наборах.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
```

```python
predicted = lin_model.predict(X_test)
expected = Y_test

plt.figure(figsize=(14, 9))
plt.scatter(expected, predicted)
plt.plot([0, 50], [0, 50], '--k')
plt.axis('tight')
plt.xlabel('True price ($1000s)', fontsize = 15)
plt.ylabel('Predicted price ($1000s)', fontsize = 15)
plt.tight_layout()
```

Оценка модели  
Мы будем оценивать нашу модель, используя RMSE и R2-score.

```python
from sklearn.metrics import r2_score
```

```python
# оценка модели для обучающего набора
y_train_predict = lin_model.predict(X_train)
rmse = (np.sqrt(mean_squared_error(Y_train, y_train_predict)))
r2 = r2_score(Y_train, y_train_predict)

print("Производительность модели для обучающей выборки")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
print("\n")

# оценка модели для набора для испытаний
y_test_predict = lin_model.predict(X_test)
rmse = (np.sqrt(mean_squared_error(Y_test, y_test_predict)))
r2 = r2_score(Y_test, y_test_predict)

print("Производительность модели для тестовой выборки")
print("--------------------------------------")
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```

R в квадрате ближе к 1, и нет большой разницы между значением R в квадрате набора Train и Test, поэтому мы можем сказать, что модель не переобучена.  

Рассматривая RMSE: мы можем сделать вывод, что эта средняя ошибка модели равна RMSE на medv, что означает RMSE * 1000 в деньгах.

```python
print('MAE:', mean_absolute_error(Y_train, y_train_predict))
print('MSE:', mean_squared_error(Y_train, y_train_predict))
```

```python
print('MAE:', mean_absolute_error(Y_test, lin_model.predict(X_test)))
print('MSE:', mean_squared_error(Y_test, lin_model.predict(X_test)))
```

По результатам наша модель имеет точность всего 68,23% обучающей выборки и 73.61 % тестовой выборки. Итак, подготовленная модель не очень подходит для прогнозирования цен на жилье.
