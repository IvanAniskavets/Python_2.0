# Задание 1

```python
data_games = pd.read_csv(os.path.join('C:', os.sep, 'Netology', 'video_games_sales.csv'))
data_games.head()
```

## Пункт 1

```python
data_score_sports = data_games[['Genre', 'Critic_Score']][(data_games['Genre'] == 'Sports')]
```

## Как критики относятся к спортивным играм?
- H0: средняя удовлетворенность критиков спортивными играми <= 50%  
- H1: средняя удовлетворенность критиков спортивными играми  составляет не менее 50%

```python
from scipy import stats as st

alpha = 0.05
result = st.ttest_1samp(data_score_sports['Critic_Score'], 5, nan_policy='omit')

if (result.statistic > 0) & (result.pvalue / 2 < alpha):
    print('Отвергаем нулевую гипотезу, среднее больше 5')
else:
    print('Не отвергаем нулевую гипотезу, спортивные игру не нравится критикам')
```

## Пункт 2

```python
data_score_pc = data_games[['Platform', 'Critic_Score']][(data_games['Platform'] == 'PC')]
data_score_ps4 = data_games[['Platform', 'Critic_Score']][(data_games['Platform'] == 'PS4')]
```

## Критикам нравятся больше игры на PC или на PS4?
- H0: средняя оценка критиков игр на PC и PS4 одинаковая
- H1: средняя оценка критиков игр на PC и PS4 различаются

Здесь нам уже нужен двухвыборочный двухсторонний критерий.

```python
from scipy import stats as st

alpha = 0.05
result = st.ttest_ind(data_score_pc['Critic_Score'], data_score_ps4['Critic_Score'], equal_var=False, nan_policy='omit')
print(result)
if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу, критики по разному относятся к играм на PC и PS4')
else:
    print('Не отвергаем нулевую гипотезу')
```

## Пункт 3

```python
data_score_shooter = data_games[['Genre', 'Critic_Score']][(data_games['Genre'] == 'Shooter')]
data_score_strategy = data_games[['Genre', 'Critic_Score']][(data_games['Genre'] == 'Strategy')]
```

## Критикам одинаково нравится стрелялки и стратегии?
- H0: средняя оценка критиков игр стрелялок и стратегий одинаковая
- H1: средняя оценка критиков игр стрелялок и стратегий различаются

Здесь нам уже нужен двухвыборочный двухсторонний критерий.

```python
from scipy import stats as st

alpha = 0.05

result = st.ttest_ind(data_score_shooter['Critic_Score'], data_score_strategy['Critic_Score'], equal_var=False, nan_policy='omit')
print(result)
if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу, критики по разному относятся к стрелялкам и стратегиям')
else:
    print('Не отвергаем нулевую гипотезу')
```

```python
import re
```

```python
data_spam = pd.read_csv(os.path.join('C:', os.sep, 'Netology', 'spam.csv'))
data_spam.head()
```

```python
# приведение всех строк к нижнему регистру

data_spam['Message'] = data_spam.Message.str.lower()
data_spam
```

```python
# уберем все служебные символы

data_spam['Message'] = [re.sub('[\W_]+',' ', message) for message in data_spam['Message']]
data_spam
```

```python
# разделим все на слова

data_spam['Message'] = [message.split() for message in data_spam['Message']]
data_spam.head(5)
```

```python
# удалим стоп-слова

from nltk.corpus import stopwords

stopwords_set = set(stopwords.words('english'))
data_spam['Message'] = [[word for word in message if word not in stopwords_set] for message in data_spam['Message']]
data_spam.head(5)
```

Произведем лемматизацию

Лемматизация — процесс приведения словоформы к лемме — её нормальной (словарной) форме. В русском языке:

для существительных — именительный падеж, единственное число;
для прилагательных — именительный падеж, единственное число, мужской род;
для глаголов, причастий, деепричастий — глагол в инфинитиве несовершенного вида.
В других языках – по аналогии.

```python
#приведем слова к нормальной форме

from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
data_spam['Message'] = [[wordnet_lemmatizer.lemmatize(word) for word in message] for message in data_spam['Message']]


data_spam.head(5)
```

```python
data_spam['Message'] = [' '.join(message) for message in data_spam['Message']]
data_spam.head(5)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(data_spam.Message)
names = tfidf.get_feature_names()
tfidf_matrix = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names())
```

```python
tfidf_matrix
```

```python
# делим данные на тренировочные и тестовые

from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(tfidf_matrix, data_spam['Category'], test_size=0.30, random_state=42)
```

```python
# применим модель логистической регрессии 

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=42)
```

```python
lr.fit(X_test, y_test)
```

```python
lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

y_pred = lr.predict(X_test)

result = confusion_matrix(y_test, y_pred)

result
```

```python
accuracy = (result[0][0] + result[0][1]) / (result[0][0] + result[0][1] + result[1][0] + result[1][1])
recall = result[0][0] / (result[0][0] + result[1][0])
precision = result[0][0] / (result[0][0] + result[1][0])
specificity = result[1][1] / (result[1][0] + result[1][1])
f_score = (2 * recall * precision) / (recall + precision)

print('Точность классификации равна ', accuracy, '.')
print('Чувствительность ака равна',  recall, '.')
print('Точность равна ',  precision, '.')
print('Специфичность равна ',  specificity, '.')
print('F-оценка равна ',  f_score, '.')
```
